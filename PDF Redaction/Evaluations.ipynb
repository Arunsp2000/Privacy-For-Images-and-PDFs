```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
inputPath = '/content/drive/MyDrive/similarities.csv'
df = pd.read_csv(inputPath)
```

```python
descriptive_stats = df['Similarities'].describe()

# Print the descriptive statistics
print(descriptive_stats)

threshold = 0.50
countAboveThreshold = 0

for i in df['Similarities']:
  if i>=threshold:
    countAboveThreshold+=1

print("The percentage above the threshold of 0.5 is:", (countAboveThreshold/len(df['Similarities'])) * 100)
```

```python
counts, bin_edges, _ = plt.hist(df['Similarities'], bins=10, edgecolor='black')
plt.xlabel('Similarity Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Similarity Scores')
plt.show()
```

```python
for i in range(len(counts)):
    print(f"Bin {i + 1}: {counts[i]} counts, Range: ({bin_edges[i]:.2f}, {bin_edges[i + 1]:.2f})")
```
